```python
import pandas as pd 
ex = pd.read_excel("winterriver.xlsx")
ex.head()
```

```python
ex= ex[["Date","Level(m)","bilstm","gru","pred_prophet"]]
ex.head()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
sns.set_context('poster')
rcParams['font.size'] = 100
rcParams['font.family'] = "serif"
```

```python
#ax=plt.axes()
#ax.set_facecolor("Black")
plt.plot(ex["Date"],ex["Level(m)"], label = "Original",color="blue")
plt.plot(ex["Date"],ex["bilstm"] , label = "BiLSTM",color="brown",linestyle="--",lw=8)
plt.plot(ex["Date"],ex["gru"], label = "GRU",color="yellow",linestyle=":",lw=8)
plt.plot(ex["Date"],ex["pred_prophet"], label = "FbProphet",color="darkgreen")
#plt.plot(ex["Date"],ex["pred_xg"] , label = "XGboost",color="deepskyblue")
plt.ylabel('Level (m)',size=40)
plt.xlabel('Date',size=40)
plt.xticks(rotation=0,size=29)
plt.yticks(rotation=0,size=29)
plt.title("Prediction for Winterriver watershed",size=40)
plt.legend(loc=0, prop={'size': 50})
plt.show()
```
