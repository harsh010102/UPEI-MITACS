```python
import pandas as pd 
ex = pd.read_excel("carruther_2_P.xlsx")
ex.head()
```

```python
ex= ex[["Date","Level(m)"]][0:1826]
```

```python
ex.index = pd.to_datetime(ex['Date'], format='%Y-%m-%d')
```

```python
del ex['Date']
```

```python
import seaborn as sns
sns.set_context('poster')
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 60,25
rcParams['font.size'] = 45
rcParams['font.family'] = "serif"
```

```python
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 60,25
rcParams['font.size'] = 45
rcParams['font.family'] = "serif"
train = ex[ex.index < pd.to_datetime("2017-12-31", format='%Y-%m-%d')]
test = ex[ex.index >= pd.to_datetime("2017-12-31", format='%Y-%m-%d')]

plt.plot(train, color = "black" , label="Train Set")
plt.plot(test, color = "orange", label="Test Set")
plt.ylabel('Level (m)',size=40)
plt.xlabel('Date',size=40)
plt.xticks(rotation=0,size=35)
plt.yticks(rotation=0,size=35)
plt.title("Train/Test split Carruther watershed",size=40)
plt.legend(loc=0, prop={'size': 50})
plt.show()
```
